#Método de Galerkin

Este es un tipo de método de residuos pesados, donde las funciones de peso coinciden con las funciones base $\beta_i$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x') # variable
u = sp.Function('u') # función
```

```python
ode = u(x).diff(x,x) + u(x) + x
sp.dsolve(ode,u(x))
```

```python
sol = sp.dsolve(ode, ics={u(0):0,u(1):0}) # ics es un diccionario
```

```python
sol
```

```python
u_expr = sol.rhs

u_num = sp.lambdify(x,u_expr,"numpy")

xx = np.linspace(0,10,400)
yy = u_num(xx)

plt.figure(figsize=(8, 5))
plt.plot(xx, yy, label="Solución $u(x)$", linewidth=2)
plt.title('Solución del problema de contorno\n $u^{´´} + u=-x$, $u(0)=0$ $u(1)=0$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
