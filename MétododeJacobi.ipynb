```python
import numpy as np
import pandas as pd

# Definimos la matriz A y el vector b
A = np.array([[4, -1, -1, 0],
              [-1, 4, 0, -1],
              [-1, 0, 4, -1],
              [0, -1, -1, 4]], dtype=float)

b = np.array([0, 2/3, 8/9, 14/9], dtype=float)

# Tolerancia y número máximo de iteraciones
tol = 1e-6
max_iter = 100

# Condición inicial
x = np.zeros_like(b)

# Guardar resultados para la tabla
resultados = []

# Método de Jacobi
for k in range(max_iter):
    x_new = np.zeros_like(x)
    for i in range(len(b)):
        suma = 0
        for j in range(len(b)):
            if j != i:
                suma += A[i, j] * x[j]
        x_new[i] = (b[i] - suma) / A[i, i]

    # Guardar la iteración y valores
    resultados.append([k+1] + list(x_new))

    # Criterio para parar
    if np.linalg.norm(x_new - x, ord=np.inf) < tol:
        break

    x = x_new

# Crea una tabla con los resultados
columnas = ["Iteración"] + [f"x{i+1}" for i in range(len(b))]
tabla = pd.DataFrame(resultados, columns=columnas)

print("Resultados del método de Jacobi:")
display(tabla)
```
