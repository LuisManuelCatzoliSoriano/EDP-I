Grafique la función $f(x')=e^{-\frac{(x'-0.5)^2}{0.01}}$ y suponga que esta representa una perturbación en una cuerda.

$(a)$ Mostrar que al hacer el cambio de variable $x'=x-ct$ la perturbación se desplaza a la derecha.

$(b)$ Mostrar que al hacer el cambio de variable $x'=x+ct$ la perturbación se desplaza a la derecha.

$(c)$ Graficar la superposición de las ondas viajeras de los incisos $a)$ y $b)$

Primero asignamos los valores $c=1$ y $t=0.3$

```python
import numpy as np
import matplotlib.pyplot as plt


c = 1
t = 0.3
x = np.linspace(0, 1, 500)
```

Definimos las funciones involuvradas en nuestra gráfica

```python
def f(xp):
    return np.exp(-((xp - 0.5)**2) / 0.01)

f_original = f(x)
f_der = f(x - c * t)
f_izq = f(x + c * t)
f_super = f_der + f_izq
```

Hacemos la gráfica del inciso $(a)$

```python
plt.figure(figsize=(10, 6))
plt.plot(x, f_der, label='f(x - ct) Derecha', color='blue')
plt.title('Ondas viajeras')
plt.xlabel('x')
plt.ylabel('f(x, t)')
plt.legend()
plt.grid(True)
plt.show()
```

Hacemos la gráfica del inciso $(b)$

```python
plt.figure(figsize=(10, 6))
plt.plot(x, f_izq, label='f(x + ct) Izquierda', color='red')
plt.title('Ondas viajeras')
plt.xlabel('x')
plt.ylabel('f(x, t)')
plt.legend()
plt.grid(True)
plt.show()
```

Hacemos la gráfica del inciso $(c)$ con la función inicial, cuando va a la derecha, a la izquierda y finalmente la superposición

```python
plt.figure(figsize=(10, 6))
plt.plot(x, f_original, label='f(x) (Original)', color='black', linestyle='--')
plt.plot(x, f_der, label='f(x - ct) (Derecha)', color='blue')
plt.plot(x, f_izq, label='f(x + ct) (Izquierda)', color='red')
plt.plot(x, f_super, label='Superposición', color='green')
plt.title('Ondas viajeras')
plt.xlabel('x')
plt.ylabel('f(x, t)')
plt.legend()
plt.grid(True)
plt.show()

```
