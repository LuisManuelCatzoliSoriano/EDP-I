Serie de Fourier de $f(x)= \left\{\begin{matrix}
0 \quad si \quad -2 \leq x<0 \\ 2-x \quad si \quad 0 \leq x \leq 0
\end{matrix}\right.$

Definimos la serie de Fourier que calculamos:

```python
def u(x,N=20): # N es el número de armónicos
    f = np.zeros(len(x))
    g = np.zeros(len(x))
    for n in range(1,N):
        g += (2*(1+(-1)**(n+1))/(n*np.pi)**2)*np.cos(n*np.pi*x/2)+(2/(n*np.pi))*np.sin(n*np.pi*x/2)
        f=1/2+g
    return f
```

Graficamos $f(x)$ y tambièn nuestra aproximaciòn con la serie de Fourier usando disntintos armònicos

```python
x1=np.linspace(-2,0,200)
x2=np.linspace(0,2,200)
x3=np.linspace(-2,2,200)

y1=np.zeros(len(x1))
y2=2-x2
y3=u(x3,N=200)

plt.title("Gràfica de f(x)")

plt.plot(x1,y1,label='y=0')
plt.plot(x2,y2,label='y=2-x')
plt.plot(x3,y3,label='u=(N=200)')

for i in range (3,20,5):
  plt.plot(x3,u(x3,i),label='$f_N(x)$ con N=%d' %i)
  plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Variaciòn de Armonicos")

plt.legend()

plt.grid(True)
plt.show()

```
