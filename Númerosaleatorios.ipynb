```python
import tabulate
```

```python
def generador(a,c,m, x0):
    x1=(a*x0+c) %m
    return x1
```

```python
n=70
a=1
c=46
m=64
x0=1

muestra=[]
datos=[]

for i in range (n):
    nuevo=generador(a,c,m,x0)
    muestra.append([i,x0,nuevo/64])
    datos.append(x0/64)
    x0=nuevo

encabezados=["i", "x_i", "x_{i+1}"]
print(tabulate.tabulate(muestra, headers=encabezados, tablefmt="psql", numalign="center"))
```
