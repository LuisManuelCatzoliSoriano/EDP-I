```python
import tabulate
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generador(a,c,m, x0):
    x1=(a*x0+c) %m
    return x1
```

Ejercicio 1.
Tomar $m=2^7$ y elegir los parametros adecuados para obtener una muestra de numeros aleatorios con ciclo de vida completo.
a) Guardar los n√∫meros en una lista y obtener su media.
b) Graficar el histograma de la muestra.

```python
n=130
a=5 #5
c=1 #1
m=128
x0=1

muestra=[]
datos=[]

for i in range (n):
    nuevo=generador(a,c,m,x0)
    muestra.append([i,x0,nuevo/m])
    datos.append(x0/m)
    x0=nuevo

encabezados=["i", "x_i", "x_{i+1}"]
print(tabulate.tabulate(muestra, headers=encabezados, tablefmt="psql", numalign="center"))
```
