Prueba de los Promedios

Usaremos $H_0:\mu=0.5$

Estadístico de prueba:

$\bar{x}=\frac{u_1+u_2+...+u_n}{N}$

$Z_0=\frac{(\bar{x}-0.5)\sqrt{N}}{\sqrt{\frac{1}{12}}}$

Si $|Z_0|<Z_{\frac{\alpha}{2}}$, entonces no se rechaza la hipótesis. Con $\alpha$ siendo el nivel de confianza.

Esta prueba se usa para saber si un conjunto de datos proviene de la distribución uniforme.

```python
def pdpromedios(data, alpha):
    x = np.mean(data)
    z_0 = (x - 0.5) * np.sqrt(100) / np.sqrt(1/12)
    z_1 = np.abs(z_0)
    z = stats.norm.ppf(1 - alpha/2)
    if z_1 < z:
        return "No se rechaza la hipótesis"
    else:
        return "Se rechaza la hipótesis"
```

```python
pdpromedios(data, 0.05)
```
